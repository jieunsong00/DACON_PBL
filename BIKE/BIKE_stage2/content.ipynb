# 7. 평균값으로 submission 데이터 채우기

[문제 6]  
submission 데이터의 number_of_rentals 피처(feature)에 mean_target 값을 할당해 보세요.

````python
submission['number_of_rentals'] = ___
submission.head()
````

````python
#checkcode
ensure_vals(globals(),'submission','mean_target')
@check_safety
def check(
    user_submission = submission, 
    variable = 'number_of_rentals',
    user_check = mean_target
):

    c_point0 = (user_submission[variable] == user_check).all()
    if c_point0:
        return True
    else:
        return False
check()
````

### Inst.
이번 프로젝트는 따릉이 대여량을 예측한 값으로 submission 데이터를 채우는 것이 목표라고 했었는데요.  
이번 스텝에서는 따릉이 대여량을 예측한 값으로 채우지 않고,   
train 데이터셋에서 구한 number_of_rentals의 평균값으로 채워보려고 합니다.  

train 데이터셋에서 구한 number_of_rentals의 평균값으로 채우는 이유는,  
이전 스텝에서 number_of_rentals 값이 모두 0으로 되어 있는 것을 확인했기 때문입니다.  
train 데이터셋의 number_of_rentals 열이 모두 0으로 되어 있는 경우,  
이를 그대로 제출하면 예측 성능 점수가 매우 낮게 나올 가능성이 있습니다.  

이를 방지하기 위해서는 number_of_rentals 열을 합리적으로 예측할 수 있는 다른 값을 채워서 제출해야 합니다.  
일반적으로, number_of_rentals 열의 평균값(mean)을 구해서 이를 채우는 방법이 많이 사용됩니다.  
이렇게 하면 최소한의 예측 성능은 보장됩니다.  
다만, 이를 통해 높은 예측 성능을 얻기 위해서는 전처리나 모델 선택 등의 추가적인 작업이 필요합니다.  

이를 위해서는 submission 데이터의 number_of_rentals 피처(feature)에 mean_target 값을 할당하면 됩니다.  
이때, submission['target'] = mean_target와 같이 값을 할당할 수 있습니다.  

>- submission['특성 이름'] = 대체할 값

### Hint.
empty

### Solution.
```python
submission['number_of_rentals'] = mean_target
submission.head()
```

파일을 제출하기 위해서는 저장을 해야 합니다.  
빈칸을 채워 submission.csv 파일로 저장해 보세요.

````python
submission.___('submission.csv', index=False)
````

````python
#checkcode
ensure_vals(globals(), 'submission')
@check_safety
def check(
    user_answer=submission
):
    answer = pd.read_csv('submission.csv')

    sub_medv = user_answer['number_of_rentals'].apply(lambda x: int(x*100))
    ans_medv = answer['number_of_rentals'].apply(lambda x: int(x*100))

    c_point0 = (sub_medv == ans_medv).all()
    if c_point0:
        return True
    else:
        return False

check()
````
