# 6. 데이터프레임(DataFrame)을 CSV 파일로 저장하기
[문제 6]  
submission 객체의 `to_csv()` 메소드를 활용하여 파일을 저장해보세요.

````python
submission.___('submission.csv', index=False)
````

````python
#checkcode
ensure_vals(globals(), 'submission')
@check_safety
def check(
    user_answer=submission
):
    answer = pd.read_csv('submission.csv')

    sub_medv = user_answer['species'].apply(lambda x: int(x*100))
    ans_medv = answer['species'].apply(lambda x: int(x*100))

    c_point0 = (sub_medv == ans_medv).all()
    if c_point0:
        return True
    else:
        return False

check()
````

### Inst.
`to_csv('submission.csv', index=False)` 메서드는 데이터프레임(DataFrame)을 csv 파일로 저장하는 메서드입니다.  
저장할 csv파일의 이름을 지정하기 위해 메서드에 문자열 submission.csv를 입력하였고,  
인덱스를 csv 파일에 포함하지 않고 저장하기 위해 `index=False`를 사용했습니다.  
따라서 인덱스를 제외한 데이터만 저장됩니다.   

`to_csv()` 메서드는 데이터프레임을 CSV(Comma-Separated Values) 형식의 파일로 저장하는 기능을 제공합니다.    
csv 파일은 각 값이 쉼표(,)로 구분되어 있는 텍스트 파일입니다.
