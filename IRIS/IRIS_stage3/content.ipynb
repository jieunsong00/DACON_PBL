# 8. 데이터프레임(DataFrame)을 CSV 파일로 저장하기
[문제 7]  
앞에서 예측한 값을 submission의 species 피처(feature)에 넣어 봅시다.

````python
submission.___('submission.csv', index=False)
````

````python
#checkcode
ensure_vals(globals(), 'submission')
@check_safety
def check(
    user_answer=submission
):
    answer = pd.read_csv('submission.csv')

    sub_medv = user_answer['species'].apply(lambda x: int(x*100))
    ans_medv = answer['species'].apply(lambda x: int(x*100))

    c_point0 = (sub_medv == ans_medv).all()
    if c_point0:
        return True
    else:
        return False

check()
````
