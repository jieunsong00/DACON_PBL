때로는 중요하지 않아 보이는 피처가 모델 예측에 실제로 큰 영향을 미칠 수도 있습니다.

피처 선택은 데이터의 특성과 문제에 따라 다르며, 피처의 중요도 외에도 도메인 지식과 피처 간 상호작용을 고려하여 올바른 판단이 필요합니다.   
모델 성능과 안정성을 최적화하기 위해서는 데이터와 문제에 맞는 적절한 피처 선택 방법을 사용하는 것이 중요합니다.

```python
import pandas as pd

train = pd.read_csv('train.csv')  
test = pd.read_csv('test.csv')  
submission = pd.read_csv('sample_submission.csv')

    'sepal length (cm)': 'sepal_length',  
    'petal length (cm)': 'petal_length',  
    'sepal width (cm)': 'sepal_width',  
    'petal width (cm)': 'petal_width'

train = train.rename(columns=feature_names)

print('submission 데이터 개수: ', len(submission))

feature_names = [ 'petal_length', 'sepal_width', 'petal_width']
train_x = train[feature_names]
train_y = train['species']
```
