````python
#checkcode
ensure_vals(globals(),'_')
@check_safety
def check(user_answer=_,column='dir_prev_bfnum'):
    c_point0 = user_answer[column][0] == 1
    if c_point0:
        return True
    else:
        return False
check()
````

### Inst.

`isna()` 함수는 pandas 데이터프레임(DataFrame)에서 **결측값(missing value)을 찾는** 함수입니다.   
이 함수는 데이터프레임의 각 셀에 대하여 결측값 여부를 확인하며, **결측값이 존재하는 경우 True, 그렇지 않은 경우 False** 값을 반환합니다.   
이를 통해 데이터프레임에 포함된 **결측값의 위치**를 파악할 수 있습니다.

**결측값** 은 데이터셋에서 측정되지 않거나, 기록되지 않은 값으로, 이를 그대로 사용하면 모델의 성능을 저하시킬 수 있습니다.   
따라서 결측값이 있는 데이터를 사용하기 전에는 반드시 이를 **처리하는 과정** 이 필요합니다.

해당 코드는 train이라는 pandas 데이터프레임에서 **결측값 여부를 확인** 하는 코드입니다.   
이를 통해 train 데이터프레임에서 **결측값이 있는 열(column)들을 파악** 할 수 있습니다.   
이는 모델의 학습에 결측값이 영향을 미치는 경우 데이터 전처리를 수행할 필요가 있으므로,   
데이터의 결측값 여부를 파악하는 것은 데이터 분석에서 매우 중요한 단계 중 하나입니다.

````python
#checkcode
def check(user_answer=_):
    train_check = pd.read_csv('train.csv')
    train_check = train.notnull()
    
    if (user_answer == train_check).all().all():
        return True
    else:
        return False
check()
````

# 5. 피처(feature)별 결측치 개수 확인

[문제 5]  
`isnull().sum()` 메소드를 사용하여 pandas 데이터프레임에서 **결측값(missing value)의 개수**를 확인해보세요.

````python
train.___().___()
````

````python
#checkcode
ensure_vals(globals())
@check_safety
def check(user_answer='_'):
    check_train = pd.read_csv('train.csv')
    check_train = check_train.notnull().sum().sum()
    
    c_point0 = (sum(_) == check_train).all
    if c_point0:
        return True
    else:
        return False
check()
````

### Inst.

`isnull().sum()` 코드는 pandas 데이터프레임에서 **결측값(missing value)의 개수**를 확인하는 코드입니다. 
 
`isnull()` 함수를 통해 데이터프레임의 각 셀에 대해 **결측값 여부를 확인** 하고,   
이후 `sum()` 함수를 이용하여 각 **열(column)별로 결측값의 총합** 을 계산합니다.

결측값이 많은 열(column)을 찾아 결측값을 처리하거나,   
해당 열(column)을 제거하는 등의 데이터 전처리 작업을 수행할 때 유용합니다.   

---

#### 결과 해석

`dir_prev_bfnum` 열(column)에 330개의 결측치가 있음을 확인할 수 있습니다.   
이러한 결측치는 데이터 분석에서 처리해주어야 하며,   
해당 열(column)을 사용할 경우 결측치를 처리하거나 해당 열(column)을 제거하는 등의 전처리 작업이 필요합니다.
