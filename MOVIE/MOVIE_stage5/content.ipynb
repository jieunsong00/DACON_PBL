`fillna()` 메소드를 사용하여 pandas 데이터프레임(DataFrame)에서 결측치(missing value)를 0으로 채워보세요.

````python
train_fill = train.___(0)
train_fill
````

````python
#checkcode
ensure_vals(globals(),'train')
@check_safety
def check(user_answer=_, user_train=train):

    c_point0 = (user_answer.shape == (600,12))

    if c_point0 :
        return True
    else:
        return False
check()
````

### Inst.

`fillna()`는 pandas 데이터프레임(DataFrame)에서 **결측치(missing value)를 특정 값으로** 채우는 함수입니다.  
해당 코드는 train 데이터프레임에서 결측치가 있는 셀을 **0으로 채운** 새로운 데이터프레임 train_fill을 반환합니다.

결측치를 평균, 중앙값, 최빈값 등의 대표값으로 대체하거나, 특정한 값으로 채우는 방법을 선택할 수 있습니다.  

하지만, 모든 결측치를 0으로 대체하는 것이 항상 좋은 방법은 아닙니다.  
결측치 대체 방법은 데이터의 특성과 분석 목적에 따라 적절한 방법을 선택해야 합니다.

### Inst.

---

#### 결과 해석
위에서 출력된 값을 보면, 모든 피처(feature)의 결측치 개수가 **0으로 대체** 된 것을 알 수 있습니다.
