# 4. 영화 관객 수 회귀분석 모델 정의 및 학습하기

[문제 3]   
`sm.OLS.from_formula()` 메서드를 사용하여 선형 회귀 모델을 생성해 봅시다.

````python
import statsmodels.api as sm

model = sm.OLS.___(formula, data=train)
model = model.fit()
````

````python
#checkcode
ensure_vals(globals(),'model')
@check_safety
def check(user_answer=model):
    if (str(type(user_answer)) == "<class 'statsmodels.regression.linear_model.RegressionResultsWrapper'>"):
        return True
    else:
        return False
check()
````

### Inst.
해당 코드는 `statsmodels` 라이브러리(Library)를 활용하여 **회귀분석 모델을 생성하고 피팅(fitting)** 하는 과정을 나타냅니다.

파이썬에서 회귀분석은 statsmodels 라이브러리를 사용하여 수행할 수 있습니다.
`statsmodels.api` 모듈을 `sm`으로 임포트한 후, `OLS.from_formula()` 함수를 사용하여 회귀분석 모델을 생성합니다.   
from_formula() 함수를 이용하면 위에서 정의한 공식(formula)을 사용해 회귀분석을 정의할 수 있습니다.

이때, 앞서 정의한 **formula 변수** 를 사용하여 종속변수와 독립변수를 설정합니다.   
데이터는 **train 데이터프레임** 을 사용합니다.

생성된 모델을 model 변수에 저장한 후, `fit()` 메서드를 호출하여 모델을 피팅합니다.   
이는 회귀분석 모델을 주어진 데이터에 **학습** 시키는 과정을 의미합니다.   

여기서 OLS(Ordinary Least Squares)는 잔차제곱합(RSS: Residual Sum of Squares)을 최소화하는 방식입니다.

````python
#checkcode
ensure_vals(globals(),'submission','predict')
@check_safety
def check(
    user_submission = submission, 
    variable = 'box_off_num',
    user_check = predict
):

    c_point0 = (user_submission[variable] == user_check).all()
    if c_point0:
        return True
    else:
        return False
check()
````
