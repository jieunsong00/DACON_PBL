### Inst.

#### 결과 해석
`model.fit()` 을 통해  모델을 학습시키면, `model.summary()` 함수를 통해 학습 결과를 여러가지 통계 수치를 통해 확인할 수 있습니다.   

지금 단계에서는 각각의 통계 수치를 모두 이해할 필요는 없습니다.   
하지만 향후 분석 작업을 원활하게 진행하려면, 이러한 통계 수치들에 대해 조금씩 익숙해지는 것이 중요합니다.   
이를 통해 더 깊이 있는 데이터 분석 능력을 향상시킬 수 있으며, 결과 해석에 있어 더 명확한 인사이트를 얻을 수 있습니다.   
여기서는 가장 중요하면서도 대표적인 통계수치에 대하여 알아 보겠습니다. 

- **R-squared (결정 계수)** : 이 수치는 0과 1 사이의 값으로, **우리 모델이 얼마나 데이터를 잘 설명하는지** 를 나타냅니다.   
R-squared 값이 높을수록 모델이 데이터를 잘 설명하고 있다고 볼 수 있습니다.   
따라서, 우리의 목표는 R-squared 값을 높이는 것이죠.
- **coef (계수)** : 이것은 **각 피쳐의 기울기**입니다. 이를 통해 **피쳐가 타겟 변수에 얼마나 영향을 주는지** 알 수 있습니다.   
coef 값이 높을수록 그 피쳐가 타겟 변수에 미치는 영향이 큽니다.   
따라서, 해당 변수의 중요성을 평가할 수 있습니다.


### Hint.

3. `fit()` 메서드를 사용하여 모델을 학습해 주세요.
