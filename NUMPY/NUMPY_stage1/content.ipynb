### Inst.
NumPy를 사용하는 방법은 다양하지만, **가장 기본적인 방법은 NumPy 배열(array)을 생성하는 것** 입니다.

NumPy 배열을 생성하는 방법에는 여러 가지가 있습니다. 
그중 하나는 NumPy의 `np.zeros()` 함수를 사용하는 것입니다.   
이 함수는 **지정한 크기의 배열을 생성**하고, 배열의 **모든 요소를 0으로** 초기화합니다.

`zeros()` 함수의 인자로는 배열의 크기를 지정하는 **튜플(tuple)이나 정수(int)** 가 들어갈 수 있습니다.   

이 함수는 **초기값이 0으로 설정된** Numpy 배열을 생성하는 데 유용하게 사용됩니다.

이러한 방법은 데이터 분석, 머신러닝 등 다양한 분야에서 자주 사용됩니다.
