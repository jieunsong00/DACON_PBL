# 6.NumPy 배열의 축(axis)을 기준으로 최솟값(min) 구하기

[문제 3]  
twelve_array 에서 각 **수평 방향** 의 최솟값을 구해보세요.  
아래 빈칸을 채워보세요.

````python
np.___(twelve_array, axis = ___)
````

````python
#checkcode
ensure_vals(globals(),'twelve_array')
@check_safety
def check(user_answer=_):
    check1 = 3 in _
    check2 = 1 in _
    check3 = 2 in _
    if check1 and check2 and check3:
        return True
    else:
        return False
check()
````

### Inst.
`min()` 함수는 인자로 배열을 받아 해당 배열에서 가장 작은 값을 찾아 반환합니다.   
`axis`를 지정하면 지정한 축의 최솟값을 구할 수 있습니다.  
`axis=0`은 **수직** 방향, `axis=1`은 **수평** 방향입니다.  

---

#### 결과 해석

코드 실행 결과 **[3, 1, 2]** 가 반환됩니다.

### Hint.
`min()`, `axis = 1` 함수를 사용하면 수평 방향의 최솟값을 구할 수 있습니다.

### Solution.
```python
np.min(twelve_array, axis = 1)
```
