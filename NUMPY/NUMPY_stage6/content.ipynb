`cumsum()` 함수는 인자로 받은 **배열의 요소들을 누적하여 합** 을 구한 배열을 반환합니다.

세 번째 요소는 two_dimension 배열의 **첫 번째, 두 번째, 세 번째** 요소를 더한 값,  
네 번째 요소는 two_dimension 배열의 **모든 요소**를 더한 값이 됩니다.

예를 들어, 주식 시장에서 일정 기간동안 주식 가격이 어떻게 변화했는지 분석할 때,   
누적 합을 구하여 총 수익률을 계산하는 등의 용도로 활용될 수 있습니다.
