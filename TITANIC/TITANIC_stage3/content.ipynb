#### [로지스틱 회귀분석]    

분류 문제에서 주로 사용되는 알고리즘으로, 일반 선형 회귀모형을 Logit 변환해 만든 식입니다. 

$$ \log(Odds \, Ratio) = \ln(\frac{p}{1-p}) $$

로짓 변환의 결과는 x에 대한 선형 함수와 동일합니다.

$$ \log(Odds \, Ratio) = b_0 + b_1 x_1 + ... + b_n x_n $$

위 두 식을 통해 아래와 같은 식을 도출할 수 있습니다.

$$ logistic\,function = \frac{e^{(b_0 + b_1 x_1 + ... + b_n x_n)}}{1+e^{(b_0 + b_1 x_1 + ... + b_n x_n)}} $$


로지스틱 회귀분석을 그림으로 표현하면 아래와 같이 표현할 수 있습니다. x축의 범위는 $[-\infty, \infty]$ 이며 y축의 범위는 [0,1] 입니다. 그리고 하늘색 곡선이 바로 로지스틱 회귀모델입니다.

#### 결과 해석

>- **No. Observations (Number of observation)**  
    - 총 표본수, 즉 train_x의 데이터 개수를 의미합니다. 

>- **DF Residuals**  
    - 잔차의 자유도를 의미합니다.  
    잔차(residual)는 표본(sample)으로 추정한 회귀식과 실제 관측값의 차이,   
    즉 여기서는 로지스틱회귀모델에서 추정한 예측결과와 정답지와의 차이를 의미합니다. 

→ No. Observation (데이타개수)는 712개인데 잔차의 자유도는 707로 데이터 개수에서   
독립변수(피쳐) 개수 4개와 및 선형회귀에서 사용되는 절편(상수)갯수 1을 빼줘야 합니다, 

>- **DF Model**
    - 독립변수의 개수

>- **Pseudo R-squ** 
    - 로지스틱 회귀분석에서는 맥파든 의사결정계수(McFadden pseudo R square)를 사용  
    - 가장 성능이 좋을 때는 1, 나쁠 때는 0  

>- **Log-likelihood (로그 가능도)**  
    - 모델이 얼마나 데이터에 적합한지 나타내는 지표입니다.  
    - 특정 통계 모델에 의해 데이터가 생성되었을 가능성을 측정합니다.  
    - Log-likelihood가 최대가 되는 W(가중치) 값을 찾는 데 사용합니다.    

>- **LL-Null**  
    - 독립 변수가 없을 때의 Log-likelihood를 설명합니다.    
    - 보다 복잡한 모형의 적합도를 비교하기 위해 사용합니다.    
    
>- **LLR p-value (Log-Likelihood Ratio p-value)**    
    - 학습한 모델과 독립변수를 축소한 모델의 Log-Likelihood 차이로 비율을 계산한 후, 카이제곱 분포를 이용하여 p값을 계산합니다    
    - 0.05 미만일 때는 축소한 모델보다 성능이 더 좋음을 나타냅니다.  

>- **coef**     
    - 회귀 계수를 나타냅니다. 각 피처가 목표 변수에 어떤 영향을 미치는지 확인할 수 있습니다.   
    양수 값은 목표 변수와 양의 상관 관계를 나타내고, 음수 값은 음의 상관 관계를 나타냅니다.

>- **std err**    
    - 추정된 회귀 계수의 표준 오차입니다.   
    모델이 얼마나 정확한지 혹은 계수 추정치가 얼마나 부정확한지에 대한 기준으로 사용됩니다.

>- **z**   
    - 회귀 계수에 대한 검정 통계량(Z-value)이며, 검정 결과의 표준화된 값입니다.   
    회귀계수가 0에 얼마나 가까운지 알기 위해 사용됩니다. 이 값에 따라 귀무가설을 기각할지 채택할지 판단합니다.

>- **P>|z|**   
    - 각 피처에 대한 가설 검정의 p-값입니다. 피처가 무의미하게 목표 변수에 영향을 미치지 않는 귀무가설을 검정합니다.   
    주로 0.05 미만의 p-값이 얻어지면, 피처가 목표 변수에 대해 통계적으로 유의미하다고 여깁니다.

>- **[0.025 0.975]**   
    - 회귀 계수에 대한 95% 신뢰 구간입니다. 계수가 이 구간 내에서 있을 가능성이 95%에 도달하므로,   
    회귀 계수의 실제 값이 이 범위내 있을 것으로 예상할 수 있습니다.

혼동행렬은 모델의 성능을 평가할 때 사용되는 지표로, 예측값이 실제 관측값을 얼마나 정확히 예측했는지 보여 주는 행렬입니다.   
TP, TF, FP, FN(참 양성, 참 음성, 거짓 양성, 거짓 음성)의 네 가지 값을 확인할 수 있습니다.

#### 결과 해석

```python
[[TN  FP]
 [FN  TP]]
```
TN (True Negative): 실제 값이 Negative인데 예측도 Negative로 정확히 예측한 개수    
FP (False Positive): 실제 값이 Negative인데 예측은 Positive로 잘못 예측한 개수   
FN (False Negative): 실제 값이 Positive인데 예측은 Negative로 잘못 예측한 개수   
TP (True Positive): 실제 값이 Positive인데 예측도 Positive로 정확히 예측한 개수
