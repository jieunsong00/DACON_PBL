# 2. 독립 변수 분포 확인

**산점도(scatterplot)** 는 **두 개의 피처(feature) 사이의 관계**를 점으로 시각화하는 그래프입니다.   
데이터의 점은 x축과 y축으로 나타내며, 각 점은 두 피처 간의 관계를 나타냅니다.

### Inst.
seaborn의 `regplot()` 함수는 두 개의 피처(feature) 간의 선형적인 관계를 시각화하는 함수입니다.   
이 함수는 선형 회귀(Linear Regression) 모델을 이용하여 데이터의 추세선을 그리며, 이를 통해 두 피처 간의 상관관계를 파악할 수 있습니다.   
x, y 축에 해당하는 두 개의 피처를 전달받아 점들을 그래프로 표현하며,

x축은 독립 변수를, y축은 종속 변수를 나타냅니다.   
이번 스텝에서는 x=train['wind_speed'], y=train['target']로 설정되어 있으므로,   
wind_speed 변수가 독립 변수, target 변수가 종속 변수가 됩니다.

```python
sns.regplot(x=train['wind_speed'], y=train['target'])
```
이 코드는 train 데이터셋에서 wind_speed 피처와 target 피처 간의 선형적인 관계를 시각화하는 코드입니다.

# 9. train 데이터의 heatmap 그리기
히트맵을 그려 변수간의 상관관계를 확인해 봅시다.

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.heatmap(data=train.corr(), annot = True, fmt = ".2f")

plt.show()
````

````python
#checkcode
#empty
````

### Inst.
**heatmap** 은 데이터의 행렬 형태를 색상으로 나타내는 시각화 방법 중 하나입니다.        
주로 데이터의 **상관관계를 시각적으로 파악**하기 위해 사용됩니다.  

heatmap은 데이터셋에서 상관관계가 높은 피처(feature)들을 쉽게 찾을 수 있도록 도와주어,     
이를 통해 **중요한 피처(feature)들을 선택**하거나 **불필요한 피처(feature)들을 제거**할 때 유용합니다.   
또한 이를 통해 **피처(feature)들 간의 다중공선성(multicollinearity)이 있는지**도 파악할 수 있습니다.  

heatmap 함수의 파라미터는 다음과 같습니다.  
- data: heatmap을 그릴 데이터 프레임  
- annot: 각 셀에 값을 표시할지 여부를 나타내는 boolen 값  
- fmt: 각 셀에 표시될 값의 형식을 나타내는 문자열  
그 외에도 cmap, center, linewidths, linecolor 등의 파라미터를 지정할 수 있습니다.  

```python
sns.heatmap(data=train.corr(), annot=True, fmt=".2f")
```
이 코드는 train 데이터셋에서 각 피처(feature) 간의 상관관계를 시각화하는 코드입니다.   
`train.corr()`은 각 피처(feature) 간의 상관계수(correlation coefficient)를 계산하여 데이터프레임(DataFrame)으로 반환합니다.   
이를 heatmap 함수의 data 인자로 전달하여 heatmap을 그리고,     
annot=True와 fmt=".2f"를 통해 각 셀에 상관계수 값을 표시하고 소수점 둘째자리까지 표시합니다.     
이를 통해 피처(feature) 간의 상관관계를 시각적으로 파악할 수 있습니다.
