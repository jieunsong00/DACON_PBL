````python
#checkcode
ensure_vals(globals())
@check_safety
def check(user_answer = _):
    c_point0 = (len(user_answer) == 5)
    if c_point0:
        return True
    else:
        return False
check()
````

### Inst.
`head()` 메서드는 데이터프레임(DataFrame)의 첫 번째부터 지정된 개수의 행(row)을 반환하는 메서드입니다.    
`head()` 메서드는 상위 일부 행을 확인하고 데이터의 구조와 내용을 파악하는 데 사용되는데,
